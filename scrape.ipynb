```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float, Numeric
```

```python
# Extract CSVs into DataFrames
world_suicides = "Resources/1985-2016.csv"
world_suicides_df = pd.read_csv(world_suicides)
world_suicides_df = world_suicides_df.drop(world_suicides_df.columns[[7, 8, 9, 10]], axis = 1)
world_suicides_df = world_suicides_df.rename(columns = {'country': 'country_name', 'suicides_no': 'suicide_number', 'suicides/100k pop': 'suicides_per_100k'})
world_suicides_df['country_name'] = world_suicides_df['country_name'].replace({'Cabo Verde': 'Cape Verde', 'Republic of Korea': 'South Korea', 'Russian Federation': 'Russia', 'Saint Vincent and Grenadines': 'Saint Vincent and the Grenadines'})
#world_suicides_df_South_Korea = world_suicides_df.loc[world_suicides_df['country_name'] == 'South Korea']
world_suicides_df
```

```python
lat_long = "Resources/concap.csv"
lat_long_df = pd.read_csv(lat_long)
lat_long_df = lat_long_df.drop(lat_long_df.columns[[4]], axis = 1)
lat_long_df = lat_long_df.rename(columns = {'CountryName': 'country_name', 'CapitalName': 'capital_name', 'CapitalLatitude': 'capital_latitude', 'CapitalLongitude': 'capital_longitude', 'ContinentName': 'continent_name'})
lat_long_df
#lat_long_df_albania = lat_long_df.loc[lat_long_df['country_name'] == 'Albania']
#lat_long_df_albania
```
